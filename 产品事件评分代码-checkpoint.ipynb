```python
def new_code(table):
    for i in range(len(table)):
        ticker = table['TICKER'].loc[i]
        if '/' in ticker:
            tList = ticker.split('/')
            for j in range(len(tList)):
                if '.SH' in tList[j] or '.SZ' in tList[j]:
                    t = tList[j]
                    if '.' in t:
                        ttList = t.split('.')
                        tt = ttList[0]
                    break
        else:
            if '.HK' in ticker:
                continue
            ttList = ticker.split('.')
            tt = ttList[0]
        table['TICKER'].loc[i] = tt
    return table
```

```python
zhiding_new = new_code(zhiding)
```

```python
# 删除包含"HK"的new_code行
zhiding_new_drop = zhiding_new[~zhiding_new['TICKER'].str.contains('HK')]

# 重置索引
zhiding_new_drop.reset_index(drop=True, inplace=True)
zhiding_new_drop.head()
```

```python
ESG_event_file_name = 'ESG_event_0418.xlsx'

reference_file = reference.values
reference = reference_file[:, 0:2]
reference_punishment_scores = reference_file[:, -1]

data = df_combine.values
# print(data)

print('文件已读取，开始处理...')

# workbook = xlwt.Workbook(encoding='utf-8')
workbook = openpyxl.Workbook()
worksheet = workbook.create_sheet('sheet1')
del workbook['Sheet']
# worksheet = workbook.add_sheet('sheet1')
worksheet.cell(1, 1, '公司代码')
worksheet.cell(1, 2, '指标名称')
worksheet.cell(1, 3, '提取日期')
worksheet.cell(1, 4, '指标数值')
```

```python
def single_index_statistic(data_source=data, index_name='产品道德处罚1', year=2022, write_with_index=0):
    index_description = []
    punishment_score_dict = ''
    for i in range(len(reference)):
        if reference[i][0] == index_name:
            index_description = reference[i][1]
            punishment_score_dict = reference_punishment_scores[i]
    if len(index_description) == 0:
        print('指标输入有误：未能与量化方案文件包含指标匹配！')
        return

    impact_factor = ast.literal_eval(punishment_score_dict)

    # print('人机交互1： 您可以检查index description是否正确')

    def single_index_statistic_halfyear(FH=True):
        # 确定统计周期
        period_begin = str(year - 1) + '-05-01'
        period_begin = datetime.datetime.strptime(period_begin, '%Y-%m-%d')
        period_end = str(year) + '-04-30'
        period_end = datetime.datetime.strptime(period_end, '%Y-%m-%d')
        if not FH:
            period_begin = str(year - 1) + '-11-01'
            period_begin = datetime.datetime.strptime(period_begin, '%Y-%m-%d')
            period_end = str(year) + '-10-31'
            period_end = datetime.datetime.strptime(period_end, '%Y-%m-%d')

        SSGS_SET = set()
        SSGS_grade = {}
        for j in range(len(data_source)):
            # for j in range(100):
            record = data_source[j]
            SSGSDM = str(record[1])
            #SSGSDM = str(record[1]).strip()
            if SSGSDM not in SSGS_SET:
                SSGS_SET.add(SSGSDM)
                SSGS_grade[SSGSDM] = 100
            if record[0].strip() in index_description:
                # 发布和发生
                sjfb = record[2]
                if isinstance(sjfb, datetime.datetime):
                    sjfb = str(sjfb)[0:10]

                if (len(sjfb) == 0) or (len(sjfb) == 4):  # 空值或只有年份没有月份的记录无法统计
                    continue

                try:
                    # 避免某些统计的时间只有月份，没有具体日期
                    if sjfb.count('-') == 1:
                        sjfb += '-01'
                    SSFB = datetime.datetime.strptime(sjfb.strip(), '%Y-%m-%d')  # 事件发布日期
                except:
                    print(record)
                    print('-*100')
                    print('上述记录日期格式可能有错误！')
                else:
                    if (SSFB >= period_begin) and (SSFB <= period_end):  # 事件发布与发生时间均在统计时间
                        if period_end.month >= SSFB.month:
                            gap = period_end.month - SSFB.month
                        else:
                            gap = 12 + period_end.month - SSFB.month
                        weight = (100 - 90 * gap / 11) / 100  # 线性权重
                        score = impact_factor[int(record[-1])]
                        SSGS_grade[SSGSDM] += score * weight  # 更新得分

                        # if (record[1] == '600028.SH') and (record[0] in ['AntiCorruption']):
                        #     print(record, impact_factor, 'weight: ', weight, 'gap: ', gap, 'score: ', score, '结果:',
                        #           weight * score)
                        # input()
        return SSGS_grade

    ssgs_grade_FH = single_index_statistic_halfyear(FH=True)  # 上半年4/30
    ssgs_grade_LH = single_index_statistic_halfyear(FH=False)  # 下半年 10/31
    total_length = len(ssgs_grade_FH) + len(ssgs_grade_LH)

    def write_to_excel(start_index, ssgs_grade_FH, ssgs_grade_LH):
        for m, (k, v) in enumerate(ssgs_grade_FH.items()):
            idx = m + 1 + start_index
            worksheet.cell(idx + 1, 1, k)
            worksheet.cell(idx + 1, 2, index_name)
            worksheet.cell(idx + 1, 3, str(year) + '0430')
            worksheet.cell(idx + 1, 4, v)

        for n, (k, v) in enumerate(ssgs_grade_LH.items()):
            idx = n + 1 + start_index + len(ssgs_grade_FH)
            worksheet.cell(idx + 1, 1, k)
            worksheet.cell(idx + 1, 2, index_name)
            worksheet.cell(idx + 1, 3, str(year) + '1031')
            worksheet.cell(idx + 1, 4, v)

    write_to_excel(write_with_index, ssgs_grade_FH, ssgs_grade_LH)
    # print(ssgs_grade_FH, ssgs_grade_LH)
    return total_length


def single_year_different_metrics(year=2022, write_with=0):
    single_year_record_count = 0
    for name, event in reference:
        record_count = single_index_statistic(index_name=name, year=year, write_with_index=write_with)  # 7540
        write_with += record_count
        single_year_record_count += record_count
    return single_year_record_count


def main(start_year=2017, end_year=2022):
    write_with = 0
    for year in range(start_year, end_year + 1):
        record_count = single_year_different_metrics(year, write_with)
        write_with += record_count
        # print('---->',write_with)


if __name__ == '__main__':
    main()
    workbook.save(ESG_event_file_name)
```
