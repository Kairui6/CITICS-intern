```python
#去除hk股票
def ticker_clean_sup(table_0):
    for i in range(len(table_0)):
        ticker_1 = table_0["supplier_p_ticker"].loc[i]
        if ',' in ticker_1:
            tList_1 = ticker_1.split(',')
            for j in range(len(tList_1)):
                if '.SH' in tList_1[j] or '.SZ' in tList_1[j]:
                    t_1 = tList_1[j]
                    break
        else:
            if '.HK' in ticker_1:
                continue
            t_1 = ticker_1

        table_0["supplier_p_ticker"].loc[i] = t_1
        
    return table_0
```

```python
def ticker_clean_cus(table_0):
    for i in range(len(table_0)):
        ticker_2 = table_0["customer_p_ticker"].loc[i]
        
        if ',' in ticker_2:
            tList_2 = ticker_2.split(',')
            for z in range(len(tList_2)):
                if '.SH' in tList_2[z] or '.SZ' in tList_2[z]:
                    t_2 = tList_2[z]
                    break
        else:
            if '.HK' in ticker_2:
                continue
            t_2 = ticker_2

        table_0["customer_p_ticker"].loc[i] = t_2
        
    return table_0
```

```python
customer_ratio_sum = result_supplier_df_new[["reporting_year","customer_p_ticker","customer_p_cost_ratio"]].merge(industry["代码"], left_on='customer_p_ticker', right_on='代码', how='inner')
customer_ratio_sum_fin = customer_ratio_sum.groupby(["reporting_year","customer_p_ticker"]).sum().reset_index()
customer_ratio_sum_fin
```

```python
customer_ratio_sum_fin.columns = ['reporting_year', 'ticker', 'customer_p_cost_ratio']
supplier_ratio_sum_fin.columns = ['reporting_year', 'ticker', 'supplier_p_revenue_ratio']
```

```python
def merge_tables(table1, table2):
    # (customer_p_ticker和reporting_year)合并
    merged_df = pd.merge(table1, table2, 
                         on=['ticker', 'reporting_year'], 
                         how='outer')

    # 选择合并后的新表格的列
    merged_df = merged_df[['ticker', 'reporting_year', 'supplier_p_revenue_ratio', 'customer_p_cost_ratio']]

    return merged_df
```

```python
merged_data = merge_tables(customer_ratio_sum_fin, supplier_ratio_sum_fin)
merged_data
```

```python
# 按股票代码和日期排序
merged_data_o = merged_data.sort_values(by = ["ticker","reporting_year"], ascending=True)
```
